### 실습 문제

Few-shot 프롬프트를 사용하여 영화 리뷰 문장의 감성을 **긍정 / 부정 / 중립** 세 가지로 분류하세요.

1. 분류 라벨과 출력 형식을 알려주는 Few-shot 예시(2~3개)를 프롬프트에 포함하세요.
2. 아래 리뷰 3개를 각각 분류하여 결과를 출력하세요.
3. `client.responses.create()` 를 사용하고, 모델은 `gpt-5-nano` 를 사용하세요.

### 테스트 입력 예시

- `"연기는 좋았지만 줄거리가 너무 지루했다."` → 부정(또는 중립)
- `"올해 본 영화 중 단연 최고였다!"` → 긍정
- `"그냥 평범하고 무난한 영화였다."` → 중립

## 정답

```python
from dotenv import load_dotenv
load_dotenv()

from openai import OpenAI
client = OpenAI()

Model = "gpt-5-nano"
```

```python
# Few-shot 예시를 포함한 프롬프트 (라벨과 출력 형식을 안내)
few_shot = (
    "다음은 영화 리뷰를 감성(긍정/부정/중립)으로 분류한 예시입니다.\n"
    "리뷰: 배우들의 연기가 정말 감동적이었어요.\n감성: 긍정\n\n"
    "리뷰: 시간이 아까울 정도로 재미없었다.\n감성: 부정\n\n"
    "리뷰: 그럭저럭 볼 만한 무난한 영화였다.\n감성: 중립\n\n"
)

reviews = [
    "연기는 좋았지만 줄거리가 너무 지루했다.",
    "올해 본 영화 중 단연 최고였다!",
    "그냥 평범하고 무난한 영화였다.",
]

for review in reviews:
    response = client.responses.create(
        model=Model,
        instructions="당신은 영화 리뷰를 긍정, 부정, 중립 중 하나로 분류합니다. 라벨만 출력하세요.",
        input=few_shot + f"리뷰: {review}\n감성:",
    )
    print(f"리뷰: {review}")
    print(f"분류: {response.output_text}\n")
```
