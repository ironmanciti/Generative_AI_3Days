### 실습 문제

`client.responses.parse()` 와 Pydantic 모델을 사용하여, 비정형 영화 소개 문장에서 구조화된 정보를 추출하세요.

1. 다음 필드를 가진 `Movie` 모델을 정의하세요: `title`(str), `year`(int), `genres`(list[str]), `director`(str).
2. 아래 문장에서 정보를 추출하여 `Movie` 형식으로 반환하세요.
3. 결과를 출력하세요.

### 테스트 입력 예시

```text
"크리스토퍼 놀란 감독의 2010년 영화 '인셉션'은 SF, 액션, 스릴러 장르의 작품입니다."
```

→ `title='인셉션', year=2010, genres=['SF', '액션', '스릴러'], director='크리스토퍼 놀란'`

## 정답

````python
from dotenv import load_dotenv
load_dotenv()

from openai import OpenAI
client = OpenAI()

Model = "gpt-5-nano"
````

````python
from pydantic import BaseModel


class Movie(BaseModel):
    title: str
    year: int
    genres: list[str]
    director: str


text = "크리스토퍼 놀란 감독의 2010년 영화 '인셉션'은 SF, 액션, 스릴러 장르의 작품입니다."

response = client.responses.parse(
    model=Model,
    input=[
        {"role": "developer", "content": "영화 소개 문장에서 제목, 개봉연도, 장르, 감독 정보를 추출하세요."},
        {"role": "user", "content": text},
    ],
    text_format=Movie,
)

movie = response.output_parsed
print(movie)
````
