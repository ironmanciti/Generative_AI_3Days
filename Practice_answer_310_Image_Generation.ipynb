### 실습 문제

Image API(`client.images.generate`)로 자신만의 프롬프트로 이미지를 생성하고 파일로 저장하세요.

1. 원하는 장면을 묘사하는 프롬프트를 작성하세요.
2. `model="gpt-image-2"` 로 이미지를 생성하세요.
3. 반환된 base64 데이터를 디코딩하여 `.png` 파일로 저장하고 화면에 표시하세요.

### 테스트 입력 예시

- `"노을 지는 해변을 배경으로 책을 읽는 고양이, 부드러운 수채화 스타일"`

## 정답

```python
from dotenv import load_dotenv
load_dotenv()

from openai import OpenAI
client = OpenAI()
```

```python
import base64
from IPython.display import Image, display

# 1. 원하는 장면 프롬프트
prompt = "노을 지는 해변을 배경으로 책을 읽는 고양이, 부드러운 수채화 스타일"

# 2. Image API 로 이미지 생성
result = client.images.generate(
    model="gpt-image-2",
    prompt=prompt,
)

# 3. base64 디코딩 후 파일 저장 및 표시
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)

with open("my_image.png", "wb") as f:
    f.write(image_bytes)

display(Image("my_image.png"))
```
