# 이미지 생성하기

OpenAI API를 사용하면 GPT Image 또는 DALL·E 모델을 이용하여 텍스트 프롬프트에서 이미지를 생성하고 편집할 수 있습니다. 이미지 생성 기능은 두 가지 API를 통해 이용할 수 있습니다.

- image API : gpt-image-1.5, gpt-image-1, 및 gpt-image-1-mini, dall-e-3 등. 단순 이미지 생성
- responses API :
    - 다중 턴 편집 : 프롬프트를 통해 이미지를 반복적으로 고품질로 편집합니다.  
    - 유연한 입력 : 바이트 데이터뿐 아니라 이미지 파일 ID도 입력 이미지로 허용합니다.

## Image API vs Responses API 비교

| 구분 | Image API | Responses API |
|:---:|:---|:---|
| **기본 개념** | Generations / Edits 전용 엔드포인트 | 대화 플로우 안에서 built-in tool로 생성 |
| **모델 지정** | GPT Image 모델(gpt-image-2) 직접 지정 | 메인라인 모델(gpt-5.5 등)이 tool로 처리 |
| **멀티턴 편집** | ❌ 단발성 (1 프롬프트 → 1 결과) | ✅ `previous_response_id`로 반복 편집 |
| **입력 방식** | 이미지 바이트 (파일 업로드) | URL · Base64 · file ID 모두 지원 |
| **생성/편집 제어** | 엔드포인트로 구분 (generations/edits) | `action` 파라미터 (auto · generate · edit) |
| **비용 구조** | 이미지 생성 비용만 | 이미지 비용 + 메인라인 모델 토큰 과금 |
| **권장 사용처** | 단일 프롬프트 생성·편집 | 대화형으로 반복 수정하는 경험 |

> **핵심**: 이미지 자체가 목적이면 **Image API**, 대화 속에서 이미지를 다루는 것이 목적이면 **Responses API**
