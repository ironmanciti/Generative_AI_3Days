### 실습 문제

GPT-2(`gpt2-medium`)로 문장을 생성하면서, **생성 길이를 바꿔 결과를 비교**해 보세요.

1. 새로운 시작 문장(예: `"In the future, robots will"`)을 토큰화하세요.
2. `model.generate()` 를 사용해 `max_length=30` 과 `max_length=60` 두 가지로 각각 문장을 생성하세요.
3. 두 결과를 디코딩하여 출력하고, 길이에 따라 문장이 어떻게 확장되는지 비교하세요.

### 테스트 입력 예시

- 시작 문장: `"In the future, robots will"` → 짧은 생성(30 토큰)과 긴 생성(60 토큰)의 내용 차이 확인

## 정답

```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# 모델 및 토크나이저 초기화
tokenizer = GPT2Tokenizer.from_pretrained("gpt2-medium")
model = GPT2LMHeadModel.from_pretrained("gpt2-medium")
```

```python
# 1. 새로운 시작 문장 토큰화
input_text = "In the future, robots will"
input_ids = tokenizer.encode(input_text, return_tensors="pt")

# 2. 생성 길이(max_length)를 바꿔 두 번 생성하고 비교
for max_len in [30, 60]:
    output = model.generate(
        input_ids,
        max_length=max_len,
        num_return_sequences=1,
        pad_token_id=tokenizer.eos_token_id,
    )
    # 3. 디코딩하여 출력
    decoded = tokenizer.decode(output[0], skip_special_tokens=True)
    print(f"[max_length={max_len}]")
    print(decoded)
    print("-" * 60)
```
